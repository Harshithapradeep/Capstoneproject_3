```python
# Loading the csv file to pandas dataframe
import pandas as pd
import json
df=pd.read_excel(r'C:\Users\hbhat\OneDrive\Desktop\VScode\Capstoneproject3\Datset\delhi_cars.xlsx')
df
```

```python
# Data processing of unstructured data into structured data by splitting data in the new car detail column into specific columns
df2 = df.copy()
df2['new_car_detail'] = df2['new_car_detail'].apply(lambda x : dict(eval(x)))
new_car_detail = df2['new_car_detail'].apply(pd.Series)
print("Length of new_car_detail", len(new_car_detail))
new_car_detail
```

```python
#Checking for null records 
new_car_detail.isnull().sum()
```

```python

# Data processing of unstructured data into structured data by splitting data in the new car overview column into specific columns
df2 = df.copy()
df2['new_car_overview'] = df2['new_car_overview'].apply(lambda x : dict(eval(x)))
df4 = df2['new_car_overview'].apply(pd.Series)

df5 = df4.explode('top')
flattened_info =df5['top'].apply(pd.Series)

# Restructure the DataFrame to have each key as a column
car_overview = flattened_info.pivot_table(index=flattened_info.groupby('key').cumcount(), 
                            columns='key', 
                            values='value', 
                            aggfunc='first')


print("Len of result", len(car_overview))
car_overview
```

```python
# Splitting data in the new car features column into specific columns

# list to hold the feature lists
feature_lists = []
data_list = []

# Iterate through each record in the specified column containing the JSON data
for json_str in df2['new_car_feature']:

    json_str = json_str.replace("'",'"')
    # Parse the JSON
    data = json.loads(json_str)

    data_list = []
    for i in data['data']:
        data_list += i['list']
    
    # Extract feature values
    features = [item['value'] for item in data_list]
    
    # Append the feature list to the main list
    feature_lists.append(features)

# Create a new DataFrame with the extracted features
df_features = pd.DataFrame({'Features': feature_lists})

# Display the new DataFrame
print(df_features)
```

```python
# Splitting data in the new car features column into specific columns

# list to hold the feature lists
specifications_list = []
data_list = []

# Iterate through each record in the specified column containing the JSON data
for json_str in df2['new_car_specs']:

    json_str = json_str.replace("'",'"')
    # Parse the JSON
    data = json.loads(json_str)

    # Add Top Specs to data_list
    data_list += data['top']

    for i in data['data']:
        data_list += i['list']
    
    # Extract feature values
    specifications = {item['key']: item['value'] for item in data_list}
    
    # Append the feature list to the main list
    specifications_list.append(specifications)

# Create a new DataFrame with the extracted features
df_specifications = pd.DataFrame(specifications_list)

# Display the new DataFrame
print(df_specifications)
```

```python
# Adding up all data columns to singe dataframe
delhi_cars = pd.concat([new_car_detail,car_overview,df_features,df_specifications],axis=1)
col_list = delhi_cars.columns.to_list()
delhi_cars.columns
```

```python
# Loading the required data to csv file for each cities
delhi_cars.to_csv(r'C:\Users\hbhat\OneDrive\Desktop\VScode\Capstoneproject3\delhi.csv',encoding='utf-8',index=False)
```
